```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

plt.style.use('tableau-colorblind10')
```

```python
# create data frame
df = pd.read_csv("movies.csv")
df
```

```python
# find null values
df.isnull().sum()
```

```python
# check data types for all columns
df.dtypes
```

```python
# information about data
df.info()
```

```python
df
```

```python
# change data type for float to int
df['votes']= df['votes'].fillna(0).astype("int64")
df['budget']= df['budget'].fillna(0).astype("int64")
df['gross']= df['gross'].fillna(0).astype("int64")
df['runtime']= df['runtime'].fillna(0).astype("int64")
df
```

```python
# drop duplicates
df.drop_duplicates()
```

```python
# budget vs gross revenue in sctter plot
plt.scatter(x = df["budget"], y=df["gross"] )
plt.title("Budget vs Gross Revenue")
plt.xlabel("Gross Earnings")
plt.ylabel("Budget for Film")
plt.show()
```

```python
# budget vs gross revenue in using regplot
sns.regplot(x="budget",y="gross",data=df,scatter_kws={"color":"green"},line_kws={"color":"red"})
```

```python
# looking for corelation
df.corr(numeric_only=True)
```

```python
Coorelation = df.corr(numeric_only=True,method='pearson')
plt.figure(figsize=(16,8))
sns.heatmap(Coorelation, annot=True)
plt.title("Coorelation Matrix for numeric features")
plt.xlabel("Movies Feature")
plt.ylabel("Movies Feature")
plt.show()
```

```python
df_numerize = df.copy()
for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype == "object"):
        df_numerize[col_name] = df_numerize[col_name].astype("category")
        df_numerize[col_name] = df_numerize[col_name].cat.codes
df_numerize
```

```python
Coorelation = df_numerize.corr(numeric_only=True,method='pearson')
plt.figure(figsize=(14,8))
sns.heatmap(Coorelation, annot=True)
plt.title("Coorelation Matrix for numeric features")
plt.xlabel("Movies Feature")
plt.ylabel("Movies Feature")
plt.show()
```

```python
coorelation_mat = df_numerize.corr()
corelationpairs = coorelation_mat.unstack()
corelationpairs
```

```python
sorted_pairs = corelationpairs.sort_values()
sorted_pairs
```

```python
high_Corelation = sorted_pairs[(sorted_pairs)> 0.5]
high_Corelation
```
